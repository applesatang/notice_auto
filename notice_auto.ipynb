```python
#패키지 로드
import requests
from bs4 import BeautifulSoup as bs
from openpyxl import Workbook
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from datetime import datetime
import time
import re

today = datetime.today().strftime('%Y-%m-%d')
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument("--start-maximized")  # 브라우저를 최대화하여 열기

# Chrome 드라이버 생성 시에 executable_path를 지정하지 않습니다.
driver = webdriver.Chrome(options=options)


#엑셀 파일 생성
wb = Workbook()
ws1 = wb.active
ws1.title = today
ws1.append(['구청','제목','날짜','링크'])
ws1.column_dimensions['B'].width = 80
ws1.column_dimensions['C'].width = 13
ws1.column_dimensions['D'].width = 150


# 정적 크롤링 함수 선언
def st_crawiling_1(public,url,path_titles,path_date,link_url):
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    notice_titles = soup.select(path_titles)
    notice_date = soup.select(path_date)
    for x,y in zip(notice_titles,notice_date):
        titles = x.text
        links = x.attrs['href']
        date = y.text
        date_int = int(date.replace('-',''))
        if date_int >= 20230801 :
            ws1.append([public,titles,date,link_url+links])
            
# 동적 크롤링 함수선언
def dy_crawiling_1(public,path_titles,path_date,link_url):
    notice_titles = driver.find_elements(By.CSS_SELECTOR,path_titles)
    notice_date = driver.find_elements(By.CSS_SELECTOR,path_date) 
    for x,y in zip(notice_titles,notice_date) : 
        titles = x.text
        links = x.get_attribute('href')
        date = y.text
        date_int = int(date.replace('-',''))
        if date_int >= 20230801 :
            ws1.append([public,titles,date,link_url+links])
    
        
#동적 크롤링 링크 온클릭 함수
def dy_crawiling_2(public,path_titles,path_date,link_url,link_url_2):
    notice_titles = driver.find_elements(By.CSS_SELECTOR,path_titles)
    notice_date = driver.find_elements(By.CSS_SELECTOR,path_date) 
    for x,y in zip(notice_titles,notice_date) : 
        titles = x.text
        onclick = x.get_attribute('onclick')
        links = re.findall("'([^']*)'",onclick)[0]
        date = y.text
        date_int = int(date.replace('-',''))
        if date_int >= 20230801 :
            ws1.append([public,titles,date,link_url+links])
        
#관악구청 한정 함수 후보 2
# def dy_crawiling_2(public,path_titles,path_date,link_url):
#     notice_titles = driver.find_elements(By.CSS_SELECTOR,path_titles)
#     notice_date = driver.find_elements(By.CSS_SELECTOR,path_date)
#     for x,y in zip(notice_titles,notice_date) :
#         titles = x.text
#         onclick_value = x.get_attribute('onclick')
#         number = extract_number_from_doBbsFView(onclick_value)
#         links = f"https://www.gwanak.go.kr/site/gwanak/ex/bbsNew/View.do?not_ancmt_mgt_no={number}&typeCode=1"
#         date = y.text
#         ws1.append([public,titles,date,link_url+links])        
        
        
        
# '안전점검' 검색
def search(search_path):
    time.sleep(3)
    search_box = driver.find_element(By.XPATH, search_path)
    search_box.send_keys('안전점검')
    search_box.send_keys(Keys.RETURN)
    time.sleep(3)


# 강남구청
url_kn = 'https://www.gangnam.go.kr/notice/list.do?mid=ID05_0402&pgno=1&gubunfield=&deptField=BNI_DEP_CODE&deptId=&keyfield=BNI_MAIN_TITLE&keyword=%EC%95%88%EC%A0%84%EC%A0%90%EA%B2%80'
st_crawiling_1('강남구청',url_kn,'td.align-l > a', 'td:nth-child(5)','https://www.gangnam.go.kr')


# 강동구청
url_kd = 'https://www.gangdong.go.kr/web/newportal/notice/01?sc=&sv=%EC%95%88%EC%A0%84%EC%A0%90%EA%B2%80'
st_crawiling_1('강동구청',url_kd,'td.tx-lt > a','td:nth-child(5)','https://www.gangdong.go.kr')
        

# 강북구청
url_gb = 'https://www.gangbuk.go.kr/www/contents.do?key=8065'
driver.get(url_gb)
time.sleep(3)

iframe = driver.find_element(By.CSS_SELECTOR, 'iframe')  
driver.switch_to.frame(iframe)

# iframe 내부에서 요소 찾기
search('//*[@id="temp"]')
dy_crawiling_1('강북구청','td:nth-child(3) > p > a','td:nth-child(5) > a','')

# 프레임에서 빠져나오기
driver.switch_to.default_content()

#강서구청_공지사항
url_gs1 = 'https://www.gangseo.seoul.kr/gs040101'
driver.get(url_gs1)

search('//*[@id="srchText"]')

dy_crawiling_1('강서구청','td.text-align.is-left > a', 'td:nth-child(4)','')
# document.querySelector("#container > div.content-body > div > div.board-list-wrap > div.responsive-shadow > div > table > tbody > tr:nth-child(1) > td.text-align.is-left > a")
# document.querySelector("#container > div.content-body > div > div.board-list-wrap > div.responsive-shadow > div > table > tbody > tr:nth-child(1) > td:nth-child(4)")

#강서구청 고시공고
url_gs2 = 'https://www.gangseo.seoul.kr/gs040301'
driver.get(url_gs2)

search('//*[@id="srchText"]')

dy_crawiling_1('강서구청','td.text-align.is-left > a', 'td:nth-child(5)',url_gs2)

#관악구청
url_ga = 'https://www.gwanak.go.kr/site/gwanak/ex/bbsNew/List.do?typeCode=1'
driver.get(url_ga)

search('//*[@id="searchKey"]')

dy_crawiling_2('관악구청', 'td:nth-child(3) > a' , 'td:nth-child(7)','https://www.gwanak.go.kr/site/gwanak/ex/bbsNew/View.do?not_ancmt_mgt_no=','&typeCode=1')

# 광진구청
url_gj = 'https://www.gwangjin.go.kr/portal/bbs/B0000003/list.do?pageIndex=1&menuNo=200192&noticeType=&deptId=&searchCnd=1&searchWrd=%EC%95%88%EC%A0%84%EC%A0%90%EA%B2%80&sdate=&edate='
st_crawiling_1('광진구청',url_gj,'td.tit > a','td:nth-child(6)','https://www.gwangjin.go.kr')


#구로구청
url_gr = 'https://www.guro.go.kr/www/selectBbsNttList.do?key=1791&bbsNo=663&integrDeptCode=&searchCtgry=&searchCnd=SJ&searchKrwd=%EC%95%88%EC%A0%84%EC%A0%90%EA%B2%80'
st_crawiling_1('구로구청',url_gr,'td.p-subject > a','td:nth-child(5)','https://www.guro.go.kr/www')

#금천구청
url_gc = 'https://www.geumcheon.go.kr/portal/tblSeolGosiDetailList.do?key=294&rep=1&searchCnd=all&searchKrwd=%EC%95%88%EC%A0%84%EC%A0%90%EA%B2%80'
st_crawiling_1('금천구청',url_gc,'td:nth-child(3) > a','td:nth-child(5)','https://www.geumcheon.go.kr/portal')

# 노원구청
url_nw = 'https://www.nowon.kr/www/user/bbs/BD_selectBbsList.do?q_bbsCode=1003&q_bbscttSn=&q_estnColumn7=&q_estnColumn1=11&q_ntceSiteCode=11&q_clCode=0&q_rowPerPage=10&q_currPage=1&q_sortName=&q_sortOrder=&q_searchKeyTy=sj___1002&q_searchVal=%EC%95%88%EC%A0%84%EC%A0%90%EA%B2%80'
st_crawiling_1('노원구청',url_nw,'td.cell-subject > a', 'td.cell-date','')
# href 확인필요

# 도봉구청
# <iframe id="AttachFileDelFrame" 
# name="AttachFileDelFrame" 
# src="https://dssl.dobong.go.kr" 
# title="백오피스처리" 
# style="display:none;width:0px;height:0px;border:0px solid red;"></iframe>

url_db = 'https://www.dobong.go.kr/Contents.asp?code=10008773'
driver.get(url_db)
time.sleep(3)

iframe = driver.find_element(By.CSS_SELECTOR, 'iframe')  
driver.switch_to.frame(iframe)

# iframe 내부에서 요소 찾기
search('//*[@id="temp"]')
dy_crawiling_2('도봉구청','td:nth-child(3) > p > a','td:nth-child(5) > a','','')


# 프레임에서 빠져나오기
driver.switch_to.default_content()


# 동대문구청
# 동작구청
# 마포구청
# 서대문구청
# 서초구청
# 성동구청
# 성북구청
# 송파구청
# 양천구청
# 영등포구청
# 용산구청
# 은평구청
# 종로구청
# 중구청
# 중랑구청

# 브라우저 닫기
driver.quit()

# 엑셀 파일 저장
wb.save(r'C:\Users\PC\Desktop\PROJECT_01\notice_date\ ' + today + '.xlsx')

```
